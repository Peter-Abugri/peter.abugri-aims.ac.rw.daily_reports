### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---

````python


# Data for plotting
x = np.linspace(1, 10, 100)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))


# First subplot: y1 = x
axs[0, 0].plot(x, x, 'r')
axs[0, 0].set_title(r"$y1 = x$")

# Second subplot: y2 = x^2
axs[0, 1].plot(x, x**2, 'g')
axs[0, 1].set_title(r"$y2 = x^2$")

# Third subplot: y3 = x^3
axs[1, 0].plot(x, x**3, 'b')
axs[1, 0].set_title(r"$y3 = x^3$")

# Fourth subplot: y4 = x^4
axs[1, 1].plot(x, x**4, 'm')
axs[1, 1].set_title(r"$y4 = x^4$")


plt.tight_layout()

plt.show()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.

```python
# Exercise 21
```

---

````python

# Define the grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create the meshgrid
X, Y = np.meshgrid(x, y)

# Define the function z = sin(x) + cos(y)
Z = np.sin(X) + np.cos(Y)

# Create the contour plot
plt.contour(X, Y, Z, levels=10, cmap='viridis')

# Add labels and a color bar
plt.title('Contour plot of z = sin(x) + cos(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='z')

# Display the plot
plt.show()
````

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.

```python
# Exercise 26
```

---

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)  # 100 points from -5 to 5
y = np.linspace(-5, 5, 100)  # 100 points from -5 to 5
x, y = np.meshgrid(x, y)  # Create a meshgrid for 3D plotting
# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Calculate z values based on the function z = x^2 + y^2
z = x**2 + y**2


ax.plot_surface(x, y, z, cmap='plasma')  # Use a colormap for better visualization
ax.set_title('3D Plot of z = x^2 + y^2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
````

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.

```python
# Exercise 27
```

---

````python


# Define the number of variables
num_vars = 5

# Create a list of variable names
variables = ['Variable 1', 'Variable 2', 'Variable 3', 'Variable 4', 'Variable 5']

# Generate random values between 0 and 1
values = np.random.rand(num_vars)

# Repeat the first value to close the circle
values = np.concatenate((values, [values[0]]))

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Repeat the first angle to close the circle

# Create the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25)  # Fill area
ax.plot(angles, values, color='blue', linewidth=2)  # Outline the area

# Add labels for each variable
ax.set_yticklabels([])  # Remove the y-labels
ax.set_xticks(angles[:-1])  # Set the position of the labels
ax.set_xticklabels(variables)  # Label each axis

# Set the title
plt.title('Radar Chart with Random Values', size=15, color='black', weight='bold')
plt.show()
````

### Exercise 30:
Create a stem plot for the function `y = sin(x)` for `x` values between 0 and 2Ï€.

```python
# Exercise 30
```

---

````python

# Define the x values between 0 and 2Ï€
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2Ï€
plt.figure(figsize=(10, 6))
# Calculate the y values for the sine function
y = np.sin(x)


plt.stem(x, y, basefmt=" ", use_line_collection=True)

# Add title and labels
plt.title('Stem Plot of y = sin(x)', fontsize=15)
plt.xlabel('x values (radians)', fontsize=12)
plt.ylabel('y values', fontsize=12)

# Set x-ticks to show important angles
ticks = np.linspace(0, 2 * np.pi, 9)  # 9 ticks
labels = ['0', 'Ï€/4', 'Ï€/2', '3Ï€/4', 'Ï€', '5Ï€/4', '3Ï€/2', '7Ï€/4', '2Ï€']  # 9 labels

plt.xticks(ticks, labels)  # Set ticks and labels

plt.grid(True)

# Show the plot
plt.show()

````
